```python
import matplotlib.pyplot as plt
labels = ['Nokia', 'Sumsang','Apple','Lumia']
values = [10,30,40,15]
colors = ['Yellow','Green','Blue','Red']
explode = [0,0,0.3,0] # for slicing one color
plt.title('A Pie Chart')
plt.pie(values, colors=colors, labels=labels, explode=explode, shadow=True, autopct='%1.1f%%', startangle=180)
plt.axis('equal')
```

```python
# Pie Charts with pandas DataFrame
import matplotlib.pyplot as plt
import pandas as pd
data = {'series1':[1,3,4,3,5],
        'series2':[2,4,5,2,6
                   ],
        'series3':[3,2,3,1,3]
        }
df = pd.DataFrame(data)
# df
df['series2'].plot(kind='pie',figsize=(6,6),autopct='%1.1f%%')
```
